```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder
```

```python
def get_title(index):
    return dataset[dataset.index == index]["title"].values[0]
```

```python
def get_index(title):
    return dataset[dataset.title == title]["new_id"].values[0]
```

```python
dataset = pd.read_csv("G:/Software/Machine learning/Datasets/Movie Recommendation/movieRecomend.csv")
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
labelencoder_X = LabelEncoder()
dataset['new_id']= labelencoder_X.fit_transform(dataset['id'])
```

```python
dataset.head(4)
```

```python
features = ['keywords','cast','genres','director']
```

```python
for feature in features:
    dataset[feature] = dataset[feature].fillna('')
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
def combine_features(row):
    try:
        return row['keywords'] +" "+row['cast']+" "+row["genres"]+" "+row["director"]
    except:
        print ("Error:", row)
```

```python
dataset["combined_features"] = dataset.apply(combine_features,axis=1)
```

```python
cv = CountVectorizer()
```

```python
cnt_matrix = cv.fit_transform(dataset["combined_features"])
cnt_matrix
```

```python
cosine_sim = cosine_similarity(cnt_matrix) 
pd.DataFrame(cosine_sim)
```

```python
user_search = "Inception"
```

```python
movie_index = get_index(user_search)
```

```python
similar_movies =  list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_list_forMovies= sorted(similar_movies,key=lambda x:x[1],reverse=True)
```
