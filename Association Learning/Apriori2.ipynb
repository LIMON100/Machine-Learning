```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = pd.read_csv('G:/Software/Machine learning/1.1/24 Apriori/Apriori_Python/Online_Retail.csv' , engine = 'python')
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
dataset['Description'] = dataset['Description'].astype(str)
dataset['InvoiceNo'] = dataset['InvoiceNo'].astype(str)
```

```python
dataset.info()
```

```python
dataset = dataset.dropna()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
data_country = dataset[dataset['Country'] == 'Germany']
```

```python
#In this dataset example, each item in the invoice(order) is spread out over one line.So, we have to essentially compress all the line items per invoice.
data_country = data_country.groupby(["InvoiceNo","Description"])["Quantity"].sum()
```

```python
data_country
```

```python
data_country.head()
```

```python
data_country = data_country.unstack()
data_country.head()
```

```python
data_country = data_country.fillna(0)
data_country.head()
```

```python
def reduce_to_binary(qty) : 
    if qty >= 1 :
        return 1
    if qty <= 0 :
        return 0
```

```python
data_country = data_country.applymap(reduce_to_binary)
data_country.head()
```

```python
data_country.shape
```

```python
frequent_items = apriori(data_country , min_support = 0.07 , use_colnames = True)

frequest_itemsets = apriori(data_country , min_support = 0.07 , use_colnames = True)

rules = association_rules(frequent_items , metric = 'lift' , min_threshold = 1)
rules.head()
```

```python
data_country = data_country.drop(columns = ['POSTAGE'])
```
